```python
mscaler=MinMaxScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=new_data1.drop(['Injury Severity'],axis=1)
y=new_data1[['Injury Severity']]
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
X_train_norm = mscaler.fit_transform(X_train)
X_test_norm = mscaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from scipy.stats import randint
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_pred_rf= rf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
accuracy = accuracy_score(y_test, y_pred_rf)
print("Accuracy:", accuracy)
```

```python
cmatrix=confusion_matrix(y_test,y_pred_rf)
print('confusion matrix:',cmatrix)
```

```python
precision = precision_score(y_test, y_pred_rf, average='weighted')
recall = recall_score(y_test, y_pred_rf, average='weighted')
f1 = f1_score(y_test, y_pred_rf, average='weighted')
```

```python
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```
