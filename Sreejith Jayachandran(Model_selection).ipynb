```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_data=pd.read_csv(r'C:\Users\user\Downloads\titanic_dataset.csv')
```

```python
titanic_data.head()
```

```python
titanic_data.info()
```

### Preprocessing

```python
titanic_data.isna().sum()
```

```python
for i in ['Age']:
    titanic_data[i]=titanic_data[i].fillna(titanic_data[i].median())
```

```python
titanic_data.Embarked.fillna(titanic_data.Embarked.mode()[0], inplace = True)
```

```python
titanic_data.drop(['Cabin'],axis=1,inplace=True)
#dropping cabin column as there is a huge number of null columns
```

```python
titanic_data.isna().sum()
```

```python
titanic_data.drop(['Name'],axis=1,inplace=True)
```

```python
titanic_data.Ticket.nunique()
```

```python
titanic_data.drop(['Ticket'],axis=1,inplace=True)
```

```python
titanic_data.head()
```

```python
titanic_data1=pd.get_dummies(titanic_data)
```

```python
titanic_data1.head()
```

```python
X=titanic_data1.drop(['Survived'],axis=1)
Y=titanic_data1.Survived
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
```

### Knn model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model=KNeighborsClassifier()
knn_model=knn_model.fit(X_train,Y_train)
```

```python
knn_score=knn_model.score(X_test,Y_test)
```

```python
knn_score
```

### SVM

```python
from sklearn.svm import SVC
```

```python
svm_model=SVC()
```

```python
svm_model=svm_model.fit(X_train,Y_train)
```

```python
svm_score=svm_model.score(X_test,Y_test)
svm_score
```

### K-fold cross validation

```python
from sklearn.model_selection import KFold
```

```python
kfold_validator=KFold(10)
```

```python
for train_index,test_index in kfold_validator.split(X,Y):
    print('Training index',train_index)
    print('Test Index',test_index)
```

```python
from sklearn.model_selection import cross_val_score
cv_result_knn=cross_val_score(knn_model,X,Y,cv=kfold_validator)
```

```python
cv_result_knn
```

```python
cv_result_svm=cross_val_score(svm_model,X,Y,cv=kfold_validator)
cv_result_svm
```

```python
np.mean(cv_result_knn)
```

```python
np.mean(cv_result_svm)
```

### Stratified K-fold cross validation

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skfold_validator=StratifiedKFold(n_splits=10)
```

```python
for train_index,test_index in skfold_validator.split(X,Y):
    print('Training index',train_index)
    print('Test Index',test_index)
```

```python
scv_result_knn=cross_val_score(knn_model,X,Y,cv=skfold_validator)
```
